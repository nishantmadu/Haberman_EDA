**Understanding the attributes and class definition**


*   Age - Age at which the surgery is performed
*   Year - Year on which the surgery is performed
*   nodes - No.of lymph nodes/Auxillary nodes present 
*   Status - 
             *   1 - Survived more than 5 years post surgery
             *   2 - Survived less than 5 years post surgery

**Class Value Counts**

```
haberman["status"].value_counts()
```

***From the above info out of 306 patients with breast cancer about 225 of them have survived post the surgery wherein 81 of them have survived less than 5 years after the surgery is performed***

```

sns.scatterplot(x= "status",y="nodes",hue="status",data=haberman)
```

```
sns.set_style("whitegrid");
sns.pairplot(haberman,hue="status",size=3)
```

```
sns.FacetGrid(haberman, hue="status", size=5) \
   .map(sns.distplot, "nodes") \
   .add_legend();
plt.show();
```

```
sns.FacetGrid(haberman, hue="status", size=5) \
   .map(sns.distplot, "age") \
   .add_legend();
plt.show();
```

```
sns.FacetGrid(haberman, hue="status", size=5) \
   .map(sns.distplot, "year") \
   .add_legend();
plt.show();
```
